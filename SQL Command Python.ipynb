````python
#Crea un dataframe que tiene las columnas, el tipo de columna y otras caracteristicas
def describe_dataframe(df):

    description = {}
    for col_name, dtype in df.dtypes.items(): #for entre los nombres de las columnas y los tipos

      num_unique = df[col_name].nunique() #cantidad de valores unicos de la columnas
      examples = df[col_name].value_counts().index[:3].tolist() #lista de 3 ejemplos de la columna
      is_numeric = pd.api.types.is_numeric_dtype(dtype) #dice si es numerico o no
      num_null = df[col_name].isnull().sum()
      percent_null = np.round(num_null / len(df) * 100, 2)
      col_description = {
            "col_name":col_name,
            "data_type": dtype,
            "num_unique": num_unique,
            "examples": examples,
            "num_null": num_null,
            "percent_null": percent_null,
        }
      if is_numeric:
            col_description.update({
                "min": np.round(df[col_name].min(), 2),
                "max": np.round(df[col_name].max(), 2),
                "mean": np.round(df[col_name].mean(), 2),
            })
      description[col_name] = col_description
    return pd.DataFrame(description).T.fillna('')
````

# **CREATE TABLE COMMAND**

````python
def get_create_table_command(df, table_name):
  columns = []
  for col_name, dtype in df.dtypes.items():
    if dtype == "int64":
      sqlite_type = "INTEGER"
    elif dtype == "float64":
      sqlite_type = "REAL"
    else:
      sqlite_type = "TEXT"

    has_null = df[col_name].isnull().any() #Vemos si hay algun null

    if not has_null:
            sqlite_type += " NOT NULL"

    if col_name == "Id" and not has_null:
            sqlite_type += " PRIMARY KEY"

    columns.append(f"\t{col_name} {sqlite_type}")

    column_defs = ", \n".join(columns)

  return f"CREATE TABLE {table_name} (\n{column_defs}\n)"



````

**EJECUTAMOS EL COMANDO**

````python
def execute_command(conn, sql_command):
  cursor = conn.cursor() #Crea un objeto cursor
  cursor.execute(sql_command) #ejecuta el comando sql
  conn.commit() #confirma los cambio en la base de datos
````

````python
execute_command(conn, create_table)
````

````python
def execute_query(conn, query):
  cursor = conn.cursor()
  cursor.execute(query)
  result = cursor.fetchall() #recupera los datos de la query ejecutada
  column_names = [column[0] for column in cursor.description] #obtiene los nombres de las columnas a traves del description del cursor
  df = pd.DataFrame(columns=column_names, data=result)
  cursor.close() #cierra el cursor
  return df
````

````python
#Obtenemos información sobre las columnas de la tabla tb_titanic en la base de datos
execute_query(conn, "PRAGMA table_info(tb_titanic)")
````

**CREAMOS TABLA AUXILIAR**

````python
def string_to_dataframe(data):
  dict_data = {
        'Id':[],
        'description':[]
  }
  lines = data.strip().split("\n") #strip elimina los espacios en blanco y split divide segun delimitador
  for line in lines:
    _id, description = line.split("\t")
    dict_data['Id'].append(_id)
    dict_data['description'].append(description)
  df = pd.DataFrame(dict_data)

  return df
````

````python
from tqdm import tqdm

def execute_many_commands(conn, sql_commands):
    cursor = conn.cursor()
    for comm in tqdm(sql_commands):  #tqdm muestra el progreso de cada uno de los comandos
        print(comm)
        cursor.execute(comm)

    conn.commit()
````
